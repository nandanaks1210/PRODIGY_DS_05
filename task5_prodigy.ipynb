# OBJECTIVE: Analysis of Traffic Accident Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('road.csv')
df
```

```python
print("Finding the missing values in the data set:")
print(df.isnull().sum())
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.countplot(x='Day_of_week',data=df)
plt.title('Number of accidents with respect to the days of a week')
plt.show()
```

Inference from the above graph is that the most number of accidents happened on Friday

```python
sns.countplot(x='Vehicle_driver_relation',data=df)
plt.title('Number of accidents with respect to the relation of the driver to the vehicle')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='Sex_of_driver',data=df)
plt.title('Number of accidents with respect to the genders')
plt.show()
```

```python
avg_severity=df.groupby('Weather_conditions')['Accident_severity'].value_counts(normalize=True).unstack().mean(axis=1)
print("Average severity by weather condition is found to be:")
print(avg_severity)
```
